```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

#Code for model

```python
df = pd.read_csv('500hits.csv',encoding='latin-1')
```

```python
df.head(10)
```

```python
df = df.drop(columns = ['PLAYER','CS'])
```

```python
df.head()
```

```python
X = df.iloc[:,0:13]
```

```python
y = df.iloc[:,13]
```

```python
X_train, X_test, y_train, y_test = train_test_split (X,y , test_size = 0.2, random_state = 11)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
X_train = scaler.fit_transform(X_train) #80% of the data
```

```python
X_test = scaler.transform(X_test) #20% of the data
```

```python
knn =KNeighborsClassifier(n_neighbors = 8)  #n_neighbors basically checks the number of neighbors to check for the classification
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print(y_pred)
```

```python
knn.score(X_test, y_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm) #confusion matrix This is how it works here 0 = true +ve , 26 = true -ve, 67 = false +ve, 0 = false -ve 
```

```python
cr = classification_report(y_test, y_pred)
print(cr) #classification report
```
